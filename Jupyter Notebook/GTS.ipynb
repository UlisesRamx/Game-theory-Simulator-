Sea $x$ el número de escenarios posibles en el juego y $a$ el número de estrategias disponibles. Para poder crear una matriz de probabilidad que represente la ocurrencia de cada escenario $x$ al tomar ciertas estrategias $a$, se necesita encontrar una manera de determinar el número de columnas (escenarios $x$) y el número de filas (estrategias $a$) para formar la matriz.

Los datos iniciales que se proporcionan son: el número de etapas que se jugarán (también conocidas como rondas), el número de jugadores que participarán en el juego y el número de estrategias que tendrá cada jugador. Por simplicidad, asumimos que todos los jugadores cuentan con el mismo número de estrategias.

Para determinar cómo se construye este tipo de matriz, consideremos el siguiente ejemplo. Dado un juego finito de 4 rondas donde participan 3 jugadores, cada uno con dos estrategias posibles, ¿cuántas columnas (escenarios $x$) y filas (estrategias $a$) se necesitan para crear una matriz que represente el juego?

Para los escenarios $a_n$, se identifica cuántas estrategias tiene cada jugador en total a partir de un escenario. $\sum_{k=0}^{E-1} S^k = \frac{S^E - S^2}{S - 1} + 2S$, cuando $S \neq 1$.

En el caso particular de $S=1$, el total de estrategias será simplemente $S$.

Abajo se encuentra un bloque de código que permite visualizar la generación de árboles y matrices a partir de un número de rondas y estrategias de jugador.
Debido a las limitaciones técnicas, procure no usar valores demasiados grandes, o el árbol y la matriz no podrán ser legibles.


```python
import networkx as nx
from networkx.drawing.nx_pydot import write_dot, graphviz_layout
import matplotlib.pyplot as plt
import numpy as np

Rondas = 3
#Jugador = 3
Estrategias = 6

def generar_arbol(Rondas, Estrategias): 
    contador_nodos = [0]
 
    def agregar_nodos(padre, profundidad): 
        if profundidad > Rondas: 
            return 
        for _ in range(Estrategias): 
            contador_nodos[0] += 1 
            hijo = contador_nodos[0] 
            G.add_edge(padre, hijo) 
            agregar_nodos(hijo, profundidad + 1) 
 
    G.add_node(0)
    agregar_nodos(0, 1) 
    return G 


def generar_matriz(Rondas, Estrategias): 
    if Estrategias != 1:
        filas = ((Estrategias**Rondas - Estrategias**2) // (Estrategias - 1)) + 2*Estrategias
    else:
        filas = Estrategias
 
    if Estrategias != 1:
        columnas = (Estrategias**Rondas - 1) // (Estrategias - 1)
    else:
        columnas = Estrategias

    matriz = np.zeros((filas, columnas))
    return matriz
    
G = nx.DiGraph()
generar_arbol(Rondas,Estrategias)
M = generar_matriz(Rondas,Estrategias)
pos = graphviz_layout(G, prog='dot')
nx.draw(G, pos, with_labels=True, node_size=350, node_color='lightblue')
plt.title('Juego finito')
plt.show()
print(G)
print(f"Dimensiones de la matriz: {M.shape}")
print(M)
```
